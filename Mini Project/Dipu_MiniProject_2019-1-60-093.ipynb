## Maze solve using DFS (Uninformed search)

### Declare a Node Class

```python
class Node():
    def __init__(self, st, prnt, act):
        self.st = st #state
        self.prnt = prnt #parent
        self.act = act  #action  
```

## Stack Class

```python
class Stack():
    def __init__(self):
        self.topNode = []
    def add(self, node):
        self.topNode.append(node)
    def contains_state(self, st):
        return any(node.st == st for node in self.topNode)
    def empty(self):
        return len(self.topNode) == 0
    def remove(self):
        if self.empty():
            raise Exception("empty topNode")
        else:
            node = self.topNode[-1]
            self.topNode = self.topNode[:-1]
            return node
```

## Queue Class

```python
class Queue(Stack):
    def remove(self):
        if self.empty():
            raise Exception("empty topNode")
        else:
            node = self.topNode[0]
            self.topNode = self.topNode[1:]
            return node
```

## Maze Class

```python
class Maze():
    
    def __init__(self,file):
        # Reading file from file and set height and width of maze
        with open(file) as foo:
            contents = foo.read()
        # Validate start and goal
        if contents.count("o") !=1:
            raise Exception("maze must have exactly a start point")
        if contents.count("x") != 1:
            raise Exception("maze must have exactly a goal")
        # Determine height and width of maze
        contents = contents.splitlines()
        self.height = len(contents)
        self.width = max(len(line) for line in contents)
        # Keep track of walls in maze
        self.walls = []
        for i in range(self.height):
            row = []
            for j in range(self.width):
                try:
                    if contents[i][j] == "o":
                        self.start = (i, j)
                        row.append(False)

                    elif contents[i][j] == "x":
                        self.goal = (i, j)
                        row.append(False)
                    elif contents[i][j] == " ":
                        row.append(False)
                    else:
                        row.append(True)
                except IndexError:
                    row.append(False)
            self.walls.append(row)
        self.solution = None
        
    def print_maze(self):
        solution = self.solution[1] if self.solution is not None else None 
        for i, row in enumerate(self.walls):
            for j, col in enumerate(row):
                if col:
                    print("#", end="")
                elif (i, j) == self.start:
                    print("o", end="")
                elif (i,j) == self.goal:
                    print("x", end="")
                elif solution is not None and (i, j) in solution:
                    print("+", end="")
                else:
                    print(" ", end="")
            print()
        print()
        
    def neighbors(self, st):
        row, col = st
        # All possible actions for solving this maze
        candidates = [
            ("Go Up", (row-1, col)),
            ("Go Down", (row + 1, col)),
            ("Go Left", (row, col - 1)),
            ("Go Right", (row, col + 1))
        ]
        # Ensure action are valid
        result = []
        for action, (r, c) in candidates:
            try:
                if not self.walls[r][c]:
                    result.append((action, (r,c)))
            except IndexError:
                continue
        return result
    
    def solve(self):
        #explored states
        self.number_of_exp = 0
        # Initialize topNode to starting position
        start = Node(st=self.start, prnt=None, act=None)
        topNode = Stack() #DFS
        topNode.add(start)
        # Initialize an empty set that explored
        self.explored = set()
        #Keep running loop until find a solution
        while True:
            # If topNode is empty, then no path
            if topNode.empty():
                raise Exception("Solution Not Found")
            # Choose a node
            node = topNode.remove()
            self.number_of_exp += 1
            #If node is the goal, then it is solution
            if node.st == self.goal:
                actions = []
                cells = []
                # Follow parent nodes to find solutions
                while node.prnt is not None:
                    actions.append(node.act)
                    cells.append(node.st)
                    node = node.prnt
                # Reverse the solution for better understanding
                actions.reverse()
                cells. reverse()
                self.solution = (actions, cells)
                return
            #If Not the goal
            self.explored.add(node.st)
            # Add neighbors to topNode 
            for act, st in self.neighbors((node.st)):
                if not topNode.contains_state((st)) and st not in self.explored:
                    child = Node(st=st, prnt=node, act=act)
                    topNode.add(child)
```

```python
print("Maze Solver")
maze = Maze("maze1.txt")
print('MAP:')
maze.print_maze()
print('Goal Node', maze.goal)
maze.solve()
maze.print_maze()
path= maze.solution
print("States Explored:", maze.number_of_exp)
print()
print('Path: ',path[0],'\n')
print('Path points: ',path[1])
```

## Maze solve using Astar (informed search)

```python
import math
from simpleai.search import SearchProblem, astar

#This class contain the method of solveing maze
class Maze(SearchProblem):
    # Initialize constructor
    def __init__(self, board):
        self.board = board
        self.goal = (0, 0)

        for y in range(len(self.board)):
            for x in range(len(self.board[y])):
                if self.board[y][x].lower() == "o":
                    self.initial = (x, y)
                elif self.board[y][x].lower() == "x":
                    self.goal = (x, y)

        super(Maze, self).__init__(initial_state=self.initial)

    # action function
    def actions(self, state):
        actions = []
        for action in COSTS.keys():
            newX, newY = self.result(state, action)
            if self.board[newY][newX] != "#":
                actions.append(action)

        return actions

    # Update according to state
    def result(self, state, action):
        x, y = state

        if action.count("up"):
            y -= 1
        if action.count("down"):
            y += 1
        if action.count("left"):
            x -= 1
        if action.count("right"):
            x += 1

        new_state = (x, y)

        return new_state

    # Checking the goal. if we reach the goal or not
    def is_goal(self, state):
        return state == self.goal

    # Count the cos when we take the action
    def cost(self, state, action, state2):
        return COSTS[action]

    # Heuristic that we use to arrive at the solution
    def heuristic(self, state):
        x, y = state
        gx, gy = self.goal

        return math.sqrt((x - gx) ** 2 + (y - gy) ** 2)

if __name__ == "__main__":
    # The map
    MAP = """
    ##############################
    #         #              #   #
    # ####    ########       #   #
    #  o #    #              #   #
    #    ###     #####  ######   #
    #      #   ###   #           #
    #      #     #   #  #  #   ###
    #     #####    #    #  # x   #
    #              #       #     #
    ##############################
    """

    # Converting map into a list
    print(MAP)
    MAP = [list(x) for x in MAP.split("\n") if x]

    # Cost for movement
    cost_regular = 1.0
    cost_diagonal = 1.7

    # Cost values in dictionary
    COSTS = {
        "up": cost_regular,
        "down": cost_regular,
        "left": cost_regular,
        "right": cost_regular,
        "up left": cost_diagonal,
        "up right": cost_diagonal,
        "down left": cost_diagonal,
        "down right": cost_diagonal,
    }

    # Maze object
    maze = Maze(MAP)
    result = astar(maze, graph_search=True)

    # the path
    path = [x[1] for x in result.path()]

    # Print the result
    print()
    for y in range(len(MAP)):
        for x in range(len(MAP[y])):
            if (x, y) == maze.initial:
                print('o', end='')
            elif (x, y) == maze.goal:
                print('x', end='')
            elif (x, y) in path:
                print('*', end='')
            else:
                print(MAP[y][x], end='')

        print()

# Print the result
    print()
    for y in range(len(MAP)):
        for x in range(len(MAP[y])):
            if (x, y) == maze.initial:
                print('o', end='')
            elif (x, y) == maze.goal:
                print('x', end='')
            elif (x, y) in path:
                print('*', end='')
            else:
                print(MAP[y][x], end='')

        print()
```
